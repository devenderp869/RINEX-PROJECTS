```python
#2.TO KNOW NUMBER OF LOCATIONS IN DIFFERENT COUNTRIES WHERE THE PEOPLE ARE AFFECTED BY COVID
df['Location'].value_counts()
```

```python
#3 TO KNOW THE GRAPH OF COVID CASES IN THE DIFFERENT COUNTRIES
import matplotlib.pyplot as plt   
plt.figure(figsize = (45,20))  
df['Location'].value_counts().plot(kind = 'bar')
plt.xticks(rotation = 75)
```

```python
#5. DATA ANALYSIS GRAPH OF THE NUMBER OF CASES OF EACH YEAR
import seaborn as sns 
sns.countplot(x = 'Year',data = df) #5
```

```python
#6. TO KNOW WEEKLY DEATHS PER MILLION
df['Weekly Deaths per Million'].value_counts() #6
```

```python
#7.TO KNOW THE TOTAL WEEKLY DEATHS IN DIFFERENT LOCATIONS
df['Weekly Deaths'].value_counts() #7
```

```python
#8.GRAPH OF THE DENSITY OF THE WEEKLY CASES PER MILLION
sns.distplot(df['Weekly Cases per Million']) #8
```

```python
#9.TOTAL VACCINATIONS COMPLETED IN DIFFERENT LOCATIONS 
df['Total Vaccinations'].value_counts() #9
```

```python
#10. GRAPH OF THE TOTAL BOOSTER DENSITY
sns.distplot(df['Total Boosters']) #10
```

```python
#11.EXPECTED NEXT WEEK'S DEATHS OF DIFFERENT LOCATIONS
df["Next Week's Deaths"].value_counts() #11
```
